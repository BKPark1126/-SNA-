```python
import pandas as pd

df111= pd.read_csv("사이버 고객상담 기타_all(utf-8-sig).csv", encoding='utf-8-sig')
df111
```

```python


import rhinoMorph
rn = rhinoMorph.startRhino()

#stop_words = "은 이 것 등 더 를 좀 즉 인 옹 때 만 원 이때 개 일 기 시 럭 갤 성 삼 스 폰 트 드 기 이 리 폴 사 전 마 자 플 블 가 해주시 때 임"
stop_words= "은 이 것 등 더 를 좀 즉 인 옹 때 만 원 이때 개 일 기 시 럭 갤 성 삼 스 폰 트 드 기 이 리 폴 사 전 마 자 플 블 가 해주시 때 임 하 있 되 않 었 사람 지 보 때 같 년 대하 고용산 주 말 일 위하 때문 말하 가 받 그렇 알 경우 부분"
stop_words = stop_words.split(' ')
stop_words[0:10]
```

```python
import re
problem_sentences = []
for post in df111['내용']:
    problem_sentences.extend(re.split('; |\.|\?|\!', post))
#title_sentences[0:100]


problem_sentences_nouns = []

for senten in problem_sentences:
    morphed_data_sentences = [noun_problem for noun_problem in rhinoMorph.onlyMorph_list(rn, senten, pos=['NNG', 'NNP','VA'],combineN=True) if noun_problem not in stop_words]
    problem_sentences_nouns.append(morphed_data_sentences)
#sentences_nouns


problem_sentences_nouns_result = []

for i in problem_sentences_nouns:
    senten_list=[]
    for j in i:
        if len(j)== 1:
            continue
        senten_list.append(j)
    problem_sentences_nouns_result.append(senten_list)
#sentences_nouns_result

problem_result = []

for i in problem_sentences_nouns_result:
    if len(i) ==0:
        continue
    i = " ".join(i)
    problem_result.append(i)

#title_result

problem_result= ",".join(problem_result)
problem_result = problem_result.replace(","," ")
#title_result

final_problem = problem_result.split(" ")
#final_title

from collections import Counter

wordInfo_problem = Counter(final_problem)
#wordInfo_title

problem_top50 = dict(wordInfo_problem.most_common(50))
problem_top50
```

```python
problem_top30 = dict(wordInfo_problem.most_common(30))
problem_top30
```

```python
import numpy as np

# 50 단어 인접행렬
problem_adjacent_matrix_50 = np.zeros((50, 50), int)
for sentence in problem_sentences_nouns_result:
    for wi, i in problem_top50_word2id.items():
        if wi in sentence:
            for wj, j in problem_top50_word2id.items():
                if i != j and wj in sentence:
                    problem_adjacent_matrix_50[i][j] += 1
problem_adjacent_matrix_50
```

```python
import numpy as np

# 50 단어 인접행렬
problem_adjacent_matrix_30 = np.zeros((30, 30), int)
for sentence in problem_sentences_nouns_result:
    for wi, i in problem_top30_word2id.items():
        if wi in sentence:
            for wj, j in problem_top30_word2id.items():
                if i != j and wj in sentence:
                    problem_adjacent_matrix_30[i][j] += 1
problem_adjacent_matrix_30
```
