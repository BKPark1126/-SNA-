```python

from bs4 import BeautifulSoup
from selenium import webdriver
import time
import sys
import re
import math
import numpy  # pip install numpy 를 해야 사용가능함
import pandas as pd  # pip install pandas 해야 사용가능하고 numpy 가 먼저 설치되어 있어야 함
import xlwt

import random
import os   # 폴더 생성하기 위해 필요함



from urllib.request import urlopen

from selenium.webdriver.chrome.options import Options

chrome_options = Options()
chrome_options.add_argument("--headless")



query_txt = input("검색어 : ")


def scroll_down(driver):
    #driver.execute_script("window.scrollTo(500,document.body.scrollHeight);")
    driver.execute_script("window.scrollBy(0,3000);") # 한페이지 20개씩 출력값
    time.sleep(1)



driver = webdriver.Chrome("C:/temp/chromedriver_240/chromedriver.exe", options= chrome_options)

driver.get('http://www1.president.go.kr/search')


driver.find_element_by_xpath('//*[@id="contents"]/div[2]/div/ul/li[5]/a').click()
time.sleep(1)

driver.find_element_by_xpath('//*[@id="chkTitle"]').click()
time.sleep(1)


element=driver.find_element_by_xpath('//*[@id="query"]')
element.send_keys(query_txt)

time.sleep(1)
driver.find_element_by_xpath('//*[@id="login"]/li[2]/input').click()
time.sleep(1)



제목=[]
청원시작일= []
청원마감일 = []
청원인원 = []
링크= []
내용 =[]




for p in range(1, 23):
    
    for i in range(0,10): 
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        #title = soup.find('#contents > div.search_PG_contens > div > div.PG_contens_pd > div:nth-child(%s) > a > h2'%(i+3)).text()
        title = soup.find_all("div", class_="PG_contens_pd_list")[i].find("a", class_='pd_list_link').find('h2').text
        #start_time = soup.find('#contents > div.search_PG_contens > div > div.PG_contens_pd > div:nth-child(%s) > a > ul > li:nth-child(1) > span' % (i+3)).get_text()
        start_time = soup.find_all("div", class_="PG_contens_pd_list")[i].find('ul', class_='PG_peti_list').find_all('li')[0].find('span').text
        
        #end_titme = soup.find('#contents > div.search_PG_contens > div > div.PG_contens_pd > div:nth-child(%s) > a > ul > li:nth-child(2) > span' % (i+3)).get_text()
        end_time = soup.find_all("div", class_="PG_contens_pd_list")[i].find('ul', class_='PG_peti_list').find_all('li')[1].find('span').text
        
        #people_num = soup.find('#contents > div.search_PG_contens > div > div.PG_contens_pd > div:nth-child(%s) > a > ul > li:nth-child(3) > span' % (i+3)).get_text()        
        people_num = soup.find_all("div", class_="PG_contens_pd_list")[i].find('ul', class_='PG_peti_list').find_all('li')[2].find('span').text
        link1 = soup.find_all("a", class_="pd_list_link")[i].get('href')
        link = "http://www1.president.go.kr"+ str(link1)
        
        driver.find_element_by_css_selector('#contents > div.search_PG_contens > div > div.PG_contens_pd > div:nth-child(%s) > a > h2'% (i+3)).click()
        time.sleep(1.5)
        
        soup2 = BeautifulSoup(driver.page_source, 'html.parser')

        content = soup2.find("div", class_="View_write").get_text()
        
        driver.back()
        time.sleep(1.5)
        
        print("제목 : ", title)
        print("청원시작일 : ", start_time, "   청원마감일 : ", end_time, "   청원인원: ", people_num)
        print("링크 : ", link)
        print("내용: " , content)
        print()
        print()
        print("="*150)
        
        제목.append(title)
        청원시작일.append(start_time)
        청원마감일.append(end_time)
        청원인원.append(people_num)
        링크.append(link)
        내용.append(content)
    
        time.sleep(1.5)
    print('='*150)
    print("                                  %s 페이지 완료                             "% (p))
    print()
    print('='*150)
    
    if p % 7 ==0:
        driver.find_element_by_xpath('//*[@id="contents"]/div[2]/div/div[3]/div/div[3]/a').click()
        time.sleep(2)
    else:
        driver.find_element_by_link_text(str(p+1)).click()
        time.sleep(2)
```
