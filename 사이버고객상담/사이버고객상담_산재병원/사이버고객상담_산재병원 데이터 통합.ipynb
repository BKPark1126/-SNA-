```python
import pandas as pd 

df1 = pd.read_csv("근복_사이버고객상담_산재병원(앞)_(utf-8-sig)_result.csv", encoding='utf-8-sig')
df1
```

```python
df2 = pd.read_csv("근복_사이버고객상담_산재보험(뒤)_(utf-8-sig)_result.csv", encoding='utf-8-sig')
df2
```

```python
DF = pd.concat([df1,df2])
DF
```

```python
DF.to_csv("사이버고객상담_산재병원_result(utf-8-sig).csv", encoding='utf-8-sig')
```

```python
DF.to_csv("사이버고객상담_산재병원_result(cp949).csv", encoding='cp949')
```

```python
df= DF[['글번호','공개여부','제목','등록일','작성자','처리지사','진행현황','내용', '답변자','답변내용']].copy()
df
```

```python
글번호 =[]
공개여부 =[]
제목 = []
등록일= []
처리지사 = []
진행현황 = []
내용 =[]
답변자 =[]
답변내용 = []

for i in df['글번호']:
    #i = i.replace("\u2024", "")
    글번호.append(i)
    

for i in df['공개여부']:
    i = i.replace("\u2024", "")
    i = i.replace("\u30fb", "")
    공개여부.append(i)
    

for i in df['제목']:
    i = i.replace("\u2024", "")
    i = i.replace("\u30fb", "")
    제목.append(i)

for i in df['등록일']:
    i = i.replace("\u2024", "")
    i = i.replace("\u30fb", "")
    등록일.append(i)

for i in df['처리지사']:
    i = i.replace("\u2024", "")
    i = i.replace("\u30fb", "")
    처리지사.append(i)

for i in df['진행현황']:
    i = i.replace("\u2024", "")
    i = i.replace("\u30fb", "")
    진행현황.append(i)

for i in df['내용']:
    i = i.replace("\u2024", "")
    i = i.replace("\u30fb", "")
    내용.append(i)

for i in df['답변자']:
    i = i.replace("\u2024", "")
    i = i.replace("\u30fb", "")
    답변자.append(i)

for i in df['답변내용']:
    i = i.replace("\u2024", "")
    i = i.replace("\u30fb", "")
    답변내용.append(i)
```

```python
df_result = df.copy()
df_result['글번호']= 글번호
df_result['공개여부'] = 공개여부
df_result['제목']= 제목
df_result['등록일'] = 등록일
df_result['처리지사'] = 처리지사
df_result['진행현황'] = 진행현황
df_result['내용'] = 내용
df_result['답변자'] = 답변자
df_result['답변내용'] = 답변내용
```

```python
df_result.to_csv("사이버고객상담_산재병원_result(cp949).csv",encoding='cp949',index= False)
```
